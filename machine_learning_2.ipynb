```python
df = df.drop(columns=['id', 'host_id', 'host_name', 'latitude', 'longitude', 
                 'reviews_por_mes','numero_de_reviews', 
                 'ultima_review', 'calculado_host_listings_count'])
df.head(5)
```

```python
quartos = { 'Shared room': 0,
           'Private room': 1,
           'Entire home/apt': 2 }
df['room_type'] = df['room_type'].map(quartos)
```

```python
df['bairro_group'].unique()
```

```python
grupos_bairros = { 'Staten Island': 0,
            'Bronx':1,
            'Queens':2,
           'Brooklyn':3 ,
           'Manhattan':4  }
df['bairro_group'] = df['bairro_group'].map(grupos_bairros)
```

```python
df.head(5)
```

```python
codificacao_bairro = df.groupby('bairro')['price'].mean().to_dict()
df['bairro_encoded'] = df['bairro'].map(codificacao_bairro)
```

Regressão Linear

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Criando estimador
modelo_regressao_linear = LinearRegression()

# Treinando usando cross validation:
scores_regressao_linear = cross_val_score(modelo_regressao_linear, X_treino, y_treino, cv=10, scoring='neg_mean_squared_error')

# Calculando a média dos MSEs negativos
mse_regressao_linear = -np.mean(scores_regressao_linear)
print("Erro Médio Quadrático Médio (MSE) médio:", mse_regressao_linear)
```

LGBMR

```python
from lightgbm import LGBMRegressor

# Criando o modelo LightGBM
modelo_lightgbm = LGBMRegressor()

# Treinando usando cross validation:
scores_lightgbm = cross_val_score(modelo_lightgbm, X_treino, y_treino, cv=10, scoring='neg_mean_squared_error')

# Calculando a média dos MSEs negativos
mse_lightgbm = -np.mean(scores_lightgbm)
print("Erro Médio Quadrático Médio (MSE) médio:", mse_lightgbm)
```

GBR

```python
from sklearn.ensemble import GradientBoostingRegressor

modelo_gbr = GradientBoostingRegressor()

scores_gbr = cross_val_score(modelo_gbr, X_treino, y_treino, cv=10, scoring='neg_mean_squared_error')

# Calculando a média dos MSEs negativos
mse_gbr = -np.mean(scores_gbr)
print("Erro Médio Quadrático Médio (MSE) médio:", mse_gbr)
```

Ridge Regression

```python
from sklearn.linear_model import Ridge

modelo_RD = Ridge()

scores_RD = cross_val_score(modelo_RD, X_treino, y_treino, cv=10, scoring='neg_mean_squared_error')

# Calculando a média dos MSEs negativos
mse_RD = -np.mean(scores_RD)
print("Erro Médio Quadrático Médio (MSE) médio:", mse_RD)
```

O modelo que teve melhor desempenho foi o lightgbm
