```python
df = df.drop(columns=['id', 'host_id', 'host_name', 'latitude', 'longitude', 
                 'numero_de_reviews', 'ultima_review','reviews_por_mes',
                 'calculado_host_listings_count'])
df.head(5)
```

```python
quartos = { 'Shared room': 0,
           'Private room': 1,
           'Entire home/apt': 2 }
df['room_type'] = df['room_type'].map(quartos)
```

```python
df['bairro_group'].unique()
```

```python
grupos_bairros = { 'Staten Island': 0,
            'Bronx':1,
            'Queens':2,
           'Brooklyn':3 ,
           'Manhattan':4  }
df['bairro_group'] = df['bairro_group'].map(grupos_bairros)
```

```python
df.head(5)
```

```python
codificacao_bairro = df.groupby('bairro')['price'].mean().to_dict()
df['bairro_encoded'] = df['bairro'].map(codificacao_bairro)
```

```python
df= df.drop(columns='bairro')
```

```python
palavras = ['Penthouse', 'Stunning', 'Townhouse', 'Luxury', 'Cozy']

for palavra_chave in palavras:
    df[palavra_chave] = df['nome'].str.contains(palavra_chave, case=False)
```

```python
df['Penthouse'].replace(False, 0, inplace=True)
df['Penthouse'].replace(True, 5, inplace=True)
df['Stunning'].replace(False, 0, inplace=True)
df['Stunning'].replace(True, 4, inplace=True)
df['Townhouse'].replace(False, 0, inplace=True)
df['Townhouse'].replace(True, 3, inplace=True)
df['Luxury'].replace(False, 0, inplace=True)
df['Luxury'].replace(True, 2, inplace=True)
df['Cozy'].replace(False, 0, inplace=True)
df['Cozy'].replace(True, 1, inplace=True)
```

```python
colunas_a_somar = ['Penthouse', 'Stunning', 'Townhouse', 'Luxury', 'Cozy']
df['Soma_Palavras_Chaves'] = df[colunas_a_somar].sum(axis=1)
```

```python
df = df.drop(columns=['nome','Penthouse', 'Stunning', 'Townhouse', 'Luxury', 'Cozy'])
```
